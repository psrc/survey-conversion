```python
# Compare new survey to old
import os
import toml
import pandas as pd
from functions import load_data, plot_display

config = toml.load(os.path.join(os.getcwd(), 'summary_configuration.toml'))
```

```python
df = load_data('_person_day.tsv', config['survey_1_dir'], config['survey_2_dir'], 
               config['survey_1_name'], config['survey_2_name'])
```

```python

df['total_tours'] = df[['retours', 'restops',
       'metours', 'mestops', 'hbtours', 'wbtours', 'uwtours']].sum(axis=1)

```

```python
plot_display(df, index='total_tours', columns='source', aggfunc='sum', values='pdexpfac')
```

```python
plot_display(df, index='total_tours', columns='source', aggfunc='count', values='pdexpfac')
```

```python
for purp in ['sc','es','pb','sh','ml','so','re','me','hb','wb','uw']:
    _df = df.pivot_table(index=purp+'tours', columns='source', aggfunc='sum', values='pdexpfac')
    _df_sum = _df.sum()
    _df = _df/_df_sum
    _df.plot(kind='barh', alpha=0.6)
```

```python
for purp in ['sc','es','pb','sh','ml','so','re','me','hb','wb','uw']:
    _df = df.pivot_table(index=purp+'tours', columns='source', aggfunc='count', values='pdexpfac')
    _df_sum = _df.sum()
    _df = _df/_df_sum
    _df.plot(kind='barh', alpha=0.6)
```

```python
for purp in ['sc','es','pb','sh','ml','so','re','me']:
    _df = df.pivot_table(index=purp+'stops', columns='source', aggfunc='sum', values='pdexpfac')
    _df_sum = _df.sum()
    _df = _df/_df_sum
    _df.plot(kind='barh', alpha=0.6)
```
